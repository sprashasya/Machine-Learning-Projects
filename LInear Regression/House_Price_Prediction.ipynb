# Improving the Model

```python
def remove_outliers_iqr(X, y):
    Q1 = X.quantile(0.25)
    Q3 = X.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    mask = ~((X < lower_bound) | (X > upper_bound)).any(axis=1)
    X_cleaned = X[mask]
    y_cleaned = y[mask]

    return X_cleaned, y_cleaned


X_cleaned, y_cleaned = remove_outliers_iqr(X, y)
```
